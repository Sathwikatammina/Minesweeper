```python
import tkinter as tk
import random
from tkinter import messagebox
from tkinter import *
from PIL import Image, ImageTk
from tkinter import *
import imageio
from PIL import Image, ImageTk
import threading
from  pygame import mixer
import os
import pygame





# Constants
FIELD_SIZE = 13
MINE_COUNT = 13

# Initialize game board and mine locations
board = [[0 for _ in range(FIELD_SIZE)] for _ in range(FIELD_SIZE)]
mine_locations = random.sample(range(FIELD_SIZE*FIELD_SIZE), MINE_COUNT)
for location in mine_locations:
    row = location // FIELD_SIZE
    col = location % FIELD_SIZE
    board[row][col] = 'X'
    for i in range(row-1, row+2):
        for j in range(col-1, col+2):
            if i >= 0 and i < FIELD_SIZE and j >= 0 and j < FIELD_SIZE and board[i][j] != 'X':
                board[i][j] += 1
                
                
                
    
video_name = "C:/Users/kedhar nadh/Downloads/New folder/mingif.mp4" #This is your video file path
video = imageio.get_reader(video_name)

def stream(label):

    for image in video.iter_data():
        frame_image = ImageTk.PhotoImage(Image.fromarray(image),size="400X200")
        label.config(image=frame_image)
        label.image = frame_image

# Function to handle button click
def click_button(button):
    if button.is_revealed:
        return
    button.is_revealed = True
    if button.is_mine:
        button.config(text='M', bg='red')
        game_over()
    else:
        row, col = button.location
        button.config(text=board[row][col])
        if board[row][col] == 0:
            for i in range(row-1, row+2):
                for j in range(col-1, col+2):
                    if i >= 0 and i < FIELD_SIZE and j >= 0 and j < FIELD_SIZE:
                        neighbor_button = window.grid_slaves(row=i, column=j)[0]
                        click_button(neighbor_button)


    
    
    
    
    
    
def game_over():
    for i in range(FIELD_SIZE):
        for j in range(FIELD_SIZE):
            button = window.grid_slaves(row=i, column=j)[0]
            if button.is_mine:
                button.config(text='M', bg='red')
            button.config(state='disabled')

    lost_window = Toplevel(window)
    lost_window.configure(bg="thistle")
    lost_window.title('Game Over')
    # Create label widget with "You Lost!!" message
    message_label = Label(lost_window, text='You Lost!!ðŸ˜ž', bg="thistle",font=('Helvetica', '24'))
    message_label.pack(padx=100, pady=100)
    
    backbutton = tk.Button(lost_window, text="EXIT",  bg="peach puff2",height=1,width=5,command=window.destroy, font=('Arial 10'))
    backbutton.place(x=160, y=150)

    
    
    
    
    
    
    
    
    

# Function to create game board
def create_board():
     global window
         window = tk.Tk()
            window.title('Minesweeper')

    for i in range(FIELD_SIZE):
        for j in range(FIELD_SIZE):
            button = tk.Button(window, width=3, height=1, font=('Helvetica', '16'))
            button.grid(row=i, column=j)
            button.location = (i, j)  # Attach button location as attribute
            button.is_mine = board[i][j] == 'X'  # Attach mine status as attribute
            button.is_revealed = False  # Attach revealed status as attribute

            # Bind click function to button
            button.config(command=lambda b=button: click_button(b))

    window.mainloop()


start_window = tk.Tk()
start_window.title('Minesweeper')

start_window.geometry("1500x1100")


my_label = tk.Label(start_window)
my_label.pack()

thread = threading.Thread(target=stream, args=(my_label,))
thread.daemon = 1
thread.start()






# Create start button
start_button = tk.Button(start_window, text="START", command=create_board,height=1,width=10, bg="deep pink", fg="black",font=('Arial 15 bold'))
start_button.image = video 

pygame.mixer.init()

pygame.mixer.music.load("C:/Users/kedhar nadh/Downloads/Gamecube Intro.mp3")
#pygame.mixer.music.play()

#start_button.play()

pygame.mixer.music.play()


#start_button.pack()
start_button.place(x=700,y=700)




def open_win():
    new = Toplevel(start_window)
    new.geometry("700x350")
    new.configure(bg='thistle3')

    Label(new, text="ðŸ‘‰ðŸ» Instructions for the game ðŸ‘ˆðŸ»", justify="center",fg="brown",bg="thistle3", font=('Helvetica 23 bold')).pack(pady=30)
   
    new.title("HELP DESK")
    Label(new, text="1.Click on the grid to open\n\n2.The number displayed on the grid shows the position of the surrounding mines\n\n3.If you accidentally click on a  mine, the game gets OVER!! ",fg="black" , bg="thistle3", justify = "left",font=('Italic 14')).pack(pady=0.5)

    
    back_button = tk.Button(new, text="BACK",  height=1,width=5,command=new.destroy)
    back_button.place(x=300, y=250)




start_button = tk.Button(start_window, text="HELP",comman=open_win,height=1,width=10, bg= "green yellow",fg="black",font=('Arial 15 bold'))
start_button.place(x=250,y=700)




start_button = tk.Button(start_window, text="QUIT",command=start_window.destroy,bg="red2",fg="black", height=1, width=10,font=('Arial 15 bold'))
start_button.place(x=1200,y=700)




start_window.mainloop()
```
